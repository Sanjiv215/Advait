#### Access Modifiers ->

```python
# Decide the scope of my data .... Decide where I can use which data
# 1. Public -> Anywhere in the code I can access the data (class where I have declared it, child class, outside)
# 2. Private -> Only the class where it declared the class can access it (not child class, not in outside)
# 3. Protected -> Only the class where it declared and the child class of it can access -> Python doesn't support protected access specifiers
# If I create a protected data it will act like public
# In Python Protected (class, child class, outside) -> same as public
```

```python
class Student1:
  def __init__(self, name, atm_pin, bank):
    self.name = name
    self.__atm_pin = atm_pin    # Private variable -> create using double underscore(__)
    self._bank = bank           # Protected variable -> create using single underscore(_)
  def __show(self):             # Private method
    print("My name is",self.name)
    print("ATM pin is", self.__atm_pin)
    print("Bank account is", self._bank)
  def private_view(self):
    self.__show()

class Teacher(Student1):
  def display(self):
    print(self.name)
    print(self._bank)
    print(self.__atm_pin)   # It will give you error
```

```python
st1 = Student1("Bhuvan", 1234, 1234567890)
```

```python
st1.name        # I can access it outside
```

```python
st1.atm_pin
```

```python
st1.__atm_pin
```

```python
st1._bank
```

```python
st1.__show()      # It become a private method
```

```python
st1.private_view()
```

## 3. Abstraction ->


---------------------------------------------------------------------------------

```python
# Abstraction -> Hiding the data -> hiding the complex part or any part that you want to hide
```

```python
# Abstract class -> It is a class that will have some abstract method
# Abstract method -> It is a method that is always empty
# All the child class of the abstract class should complete the abstract method
# Concrete class -> The child class of the abstract class where you have to complete the abstract method
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):        # It is a abstract class
  @abstractmethod         # @ -> decorator use to import or use the methods
  def sound(self):
    pass

class Dog(Animal):        # Concrete class
  def sound(self):
    print("Dog barks!!")

class Cat(Animal):        # Concrete class
  def hello(self):
    print("Cat Meow!!")
```

```python
d = Dog()
```

```python
d.sound()
```

```python
c = Cat()
```

## 4. Polymorphism ->

```python
# Polymorphism ->
# Poly -> many                Morphism -> Forms
# Many forms
```

```python
# class Myclass:
#   def add(self,a,b):
#     print(a+b)
#   def add(self, a, b, c):
#     print(a+b+c)
```

```python
# There are 2 types of Polymorphism ->
# 1. Compile time polymorphism / Method overloading ->
              # Whenever the functions are having the same name but with different number of parameters...
              # Then, At the time of calling the function the number of parameter will decide which function to call.
```

```python
class A:
  def add(self, a, b):
    print(a+b)
  def add(self, a, b, c):
    print(a+b+c)
```

```python
a = A()
```

```python
a.add(10,20)
```

```python
a.add(10,20,1)
```

```python
# 2. Run time polymorphism / Method Overriding ->
            # Whenever the functions are having the same name with same parameter, then
            # The latestfunction will override the all previous functions.
```

```python
class B:
  def add(self, a, b):
    print(a+b)
  def add(self, a, b):
    print("Hiiii")
```

```python
b = B()
```

```python
b.add(10,1)
```

```python
# Create a abstract class Shape with abstract method area()
# Create 3 child class rectangle, square, circle and complete the abstract method by calculating the areas

# time -> 9:15
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Rectangle(Shape):
  def __init__(self, length, width):
    self.length = length
    self.width = width
  def area(self):
    print(f"The area of Rectangle is: {self.length * self.width}")

class Square(Shape):
  def __init__(self, side):
    self.side = side
  def area(self):
    print(f"The area of Rectangle is: {self.side**2}")

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius
  def area(self):
    print(f"The area of Rectangle is: {3.14 * self.radius**2}")
```

```python
rec = Rectangle(10, 15)
```

```python
rec.area()
```

```python
sq = Square(25)
```

```python
sq.area()
```

```python
cir = Circle(5)
```

```python
cir.area()
```
